```python
import matplotlib.pyplot as plt
import random
import numpy as np
import pandas as pd
```

```python
from google.colab import drive

drive.mount('/content/drive')
main_directory = "/content/drive/MyDrive/PatternRecognition/googleplaystore.csv"
```

```python
df_google_play_store_apps = pd.read_csv(main_directory , nrows = 1000)
print(df_google_play_store_apps.shape)
```

```python
df_google_play_store_apps
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y) # Draw scatter plot with parameter x and y
plt.show()
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
           edgecolors="g" )#verts="<"
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
           edgecolors="g" )#verts="<"
 
plt.scatter(x,df_google_play_store_apps["Installs"], c = "y", marker = "o", s = 100, alpha=0.5, linewidths=10,
           edgecolors="c" )
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
plt.show()
```
